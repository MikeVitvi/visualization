```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

```

```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.lineplot(data = tips,      #choose our dataset
             x = 'total_bill', #define our x variable
             y = 'tip')        #define our y variable
```

```python
sns.set_style('whitegrid')

sns.lineplot(data = tips,
             x = 'total_bill',
             y = 'tip')
```

```python
tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
fig = plt.subplots(figsize = (10, 3))

tipgraph = sns.lineplot(data = tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount')
```

```python
tipgraph = sns.scatterplot(data = tips, 
                           x = 'total_bill',
                           y = 'tip',
                           style = 'time',
                           hue = 'day',
                           palette = ['purple',
                                      'hotpink',
                                      'deepskyblue',
                                      'yellowgreen'])

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tips')
```

```python
sns.pairplot(data = tips,
             hue = 'day')
```

```python
sns.relplot(data = tips,
            x = 'total_bill',
            y = 'tip',
            hue = 'sex',
            col = 'day',
            kind = 'scatter',
            col_wrap = 2)
```

```python
import plotly.graph_objects as go  # 'go' is a 'graph object'

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
graph =  go.Figure()

graph.add_trace(go.Bar(x = x1, y = y1))

graph.update_layout(title = "Pirate Scores",
                     xaxis_title = "Pirates",
                     yaxis_title = "Score")
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(
    x = x1,
    y = y1,
    mode = 'markers',
    marker = dict(
    size = 15,
    color = 'hotpink',
    opacity = 1,
    line = dict(width = 5, color = 'purple')
  ) 
))

graph.update_layout(
    title = 'Interactive Pirate Plot',
    xaxis_title = 'Pirates',
    yaxis_title = 'Scores',
    width = 500,
    height = 500
)
```

```python
from wordcloud import WordCloud

df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines = 'skip')

df

```

```python
# join all our text from each row from our quote column into a string

text = " ".join(each for each in df.quote)

# generate our wordcloude image

wordcloud = WordCloud(background_color = "white",
                      colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure

fig, ax = plt.subplots(figsize = (7, 3))
ax.imshow(wordcloud,
          interpolation = 'bilinear')

ax.axis("off")

plt.show()
```

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```
