#### Visualization 1: Distribution of Total Residential Units by Neighborhood (Python)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = '/Users/mishavitvinov/Downloads/Community Housing Data - 4326.csv'
housing_data = pd.read_csv(file_path)

# Grouping data by neighborhood number and summing total residential units
neighborhood_data = housing_data.groupby('NGHBRHD_NUM')['TTL_RES_UNIT'].sum().reset_index()

# Plotting the distribution of total residential units by neighborhood
plt.figure(figsize=(12, 8))
sns.barplot(data=neighborhood_data, x='NGHBRHD_NUM', y='TTL_RES_UNIT', hue='NGHBRHD_NUM', palette='viridis', dodge=False, legend=False)
plt.title('Distribution of Total Residential Units by Neighborhood')
plt.xlabel('Neighborhood Number')
plt.ylabel('Total Residential Units')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('/Users/mishavitvinov/Downloads/Distribution_of_Residential_Units_by_Neighborhood.png')
plt.show()

```
